# Bivariate analysis

boxplot - cat,num

jointplot- num,num
scatterplot- num,num(x)
regplot- num,num(x)

violin plot- cat,num
swarm plot- cat,num(x)

line plot- num,num

point plot- num,num

kdeplot/rugplot- num,num
rugplot - num,num(x)

barplot- cat,num


```python
for col in cat_cols:
    for ncol in num_cols:
        plt.figure(figsize=(10,3))
        plt.suptitle(f'{col}vs{ncol}')
        plt.subplot(1,3,1)    ############# 1-line,2-total graphs, 3-type of graph 
        sns.boxplot(data=data, x=col,y=ncol,palette=choice(p_list))
        plt.subplot(1,3,2)
        sns.violinplot(data=data, x=col,y=ncol,palette=choice(p_list))
        plt.subplot(1,3,3)
        sns.barplot(data=data, x=col,y=ncol,palette=choice(p_list))
        plt.show()
```

```python
  ########contrl++ctrl+d
  ##independent features
#####parameters/features affecting each other
#### Linerarity=dependency (miles,usage,features) 
##=>miles,income
pg=sns.PairGrid(data=data[num_cols],palette=choice(p_list))
pg.map_diag(plt.hist)
pg.map_lower(sns.regplot)
pg.map_upper(sns.kdeplot)

plt.show()
```

```python
sns.heatmap(data=data[num_cols].corr(),annot=True,cmap=choice(p_list))
##correlation value
```

# Multivariate analysis

```python
###cat vs cat graph:
### features= all categorical data
for col in cat_cols:
    if col=='Product':
        continue
    for ncol in num_cols:
        plt.figure(figsize=(15,5))
        plt.suptitle(f'{col}vs{ncol}')
        plt.subplot(1,3,1)    ############# 1-line,2-total graphs, 3-type of graph 
        sns.boxplot(data=data, x=col,y=ncol,palette=choice(p_list),hue='Product')
        plt.subplot(1,3,2)
        sns.violinplot(data=data, x=col,y=ncol,hue='Product',palette=choice(p_list))
        plt.subplot(1,3,3)
        sns.barplot(data=data, x=col,y=ncol,hue='Product',palette=choice(p_list))
        plt.show()
```
