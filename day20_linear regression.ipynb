## multiple linear regression


```python
target=df['Price']
features=df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
        'Area Population']] ##continuous numerical
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest,ytrain,ytest =train_test_split(features, target, test_size = 0.25,random_state=10)
```

```python
xtrain.shape  ##train
```

```python
xtest.shape  ##evaluate
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(xtrain, ytrain)
```

```python
model.score(xtest, ytest)*100
```

```python
ypred_test =model.predict(xtest)
plt.scatter(ytest, ypred_test)
plt.grid()
```

```python
query_data= {'Avg. Area Income':[200000],'Avg. Area House Age':[6],'Avg. Area Number of Rooms':[5],'Area Population':[30000]}
result=model.predict(pd.DataFrame(query_data))
print(result[0])
```
